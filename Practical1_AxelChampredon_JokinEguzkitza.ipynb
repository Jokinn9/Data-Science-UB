## 1. One dimensional case

### 1. Plot this function

Assume that $x \in R$:
$$f(x)=x^3-2x+2$$

We plot the function:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = x^3 - 2x + 2
def f(x):
    return x**3 - 2*x + 2

# Generate x values
x = np.linspace(-2, 2, 400)
y = f(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$y = x^3 - 2x + 2$', color='b')
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

As it can been seen in the graph, the function exhibits both a local maximum and a local minimum. In the following steps those points will be calculated.

### 2. Compute analytically the points

To compute the points where  $f'(x) = 0 $, we follow these steps:

1. Compute the derivative of $ f(x) = x^3 - 2x + 2 \$:

$$f'(x) = 3x^2 - 2$$

2. Set the derivative equal to zero to find the critical points:

   $$3x^2 - 2 = 0$$

3. Solve for $ x $ and therefore, the critical point are:


   $$x^2 = \frac{2}{3}
   \quad \Rightarrow \quad
   \boxed{
   x = \pm \sqrt{\frac{2}{3}} }$$


### 3. Check if the critical points are maxima or minima

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return x**3 - 2*x + 2

# Define the first derivative f'(x)
def f_prime(x):
    return 3*x**2 - 2

# Define the second derivative f''(x)
def f_double_prime(x):
    return 6*x

# Find the critical points by solving f'(x) = 0 from the previous part,

critical_points = np.array([np.sqrt(2/3), -np.sqrt(2/3)])

# Evaluate the second derivative at critical points
second_deriv_at_crit_points = f_double_prime(critical_points)

# Create the plot and mark the critical points
x_vals = np.linspace(-2, 2, 400)
f_vals = f(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, f_vals, label='$f(x) = x^3 - 2x + 2$', color='b')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()

# Mark critical points
for cp in critical_points:
    plt.scatter(cp, f(cp), color='red')
    plt.text(cp, f(cp), f' x* = {round(cp, 2)}')

plt.title('Plot of the function $f(x) = x^3 - 2x + 2$ with Critical Points')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.show()

# Output results
for i, cp in enumerate(critical_points):
    if second_deriv_at_crit_points[i] > 0:
        print(f"x* = {cp} is a minimum since f''(x*) = {second_deriv_at_crit_points[i]}")
    else:
        print(f"x* = {cp} is a maximum since f''(x*) = {second_deriv_at_crit_points[i]}")
```

### 2. Analytical Computation of the Gradient and Critical Points

Consider the function:

$$f(x_1, x_2) = x_1^2 + x_2^2$$

The gradient of a function $f(x_1, x_2)$ is given by the vector of partial derivatives:

$$\nabla f(x_1, x_2) = \left( \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2} \right)$$

Let's compute the partial derivatives of $f(x_1, x_2)$:

$$
\frac{\partial f}{\partial x_1} = \frac{\partial}{\partial x_1} \left( x_1^2 + x_2^2 \right) = 2x_1$$

$$\frac{\partial f}{\partial x_2} = \frac{\partial}{\partial x_2} \left( x_1^2 + x_2^2 \right) = 2x_2$$


Thus, the gradient is:

$$ \nabla f(x_1, x_2) = \left( 2x_1, 2x_2 \right)
$$

To find the critical points, we set the gradient equal to zero:

$$ \nabla f(x_1, x_2) = (2x_1, 2x_2) = (0, 0)
$$

This results in the following system of equations:

$$ 2x_1 = 0 \quad \text{and} \quad 2x_2 = 0
$$

Solving these equations gives:

$$ x_1 = 0 \quad \text{and} \quad x_2 = 0
$$

Therefore, the critical point is:

$$ (x_1^*, x_2^*) = (0, 0)
$$


### 3. Hessian matrix and nature of the critical point.



To determine if the critical point is a minimum, maximum, or saddle point, we compute the Hessian matrix, which contains the second-order partial derivatives.

The Hessian matrix $( H )$ is given by:
$$H = \begin{pmatrix}
\frac{\partial^2 f}{\partial x_1^2} & \frac{\partial^2 f}{\partial x_1 \partial x_2} \\
\frac{\partial^2 f}{\partial x_2 \partial x_1} & \frac{\partial^2 f}{\partial x_2^2}
\end{pmatrix}$$

Let’s compute the second-order partial derivatives:
$$\frac{\partial^2 f}{\partial x_1^2} = 2, \quad \frac{\partial^2 f}{\partial x_2^2} = 2, \quad \frac{\partial^2 f}{\partial x_1 \partial x_2} = \frac{\partial^2 f}{\partial x_2 \partial x_1} = 0
$$

Thus, the Hessian matrix is:
$$H= \begin{pmatrix}
2 & 0 \\
0 & 2
\end{pmatrix}
$$

In this particular case, $H= \nabla ^2 f(\mathbf{x}) = \nabla ^2 f(\mathbf{x}^*)$, so:

$$\nabla ^2 f(\mathbf{x}^*)= \begin{pmatrix}
2 & 0 \\
0 & 2
\end{pmatrix}$$

To check if we have a minimum or maximum, for any vector,

$$
d = \begin{pmatrix}
d_1 \\
d_2
\end{pmatrix},
$$
the product is:

$$d^T \nabla ^2 f(\mathbf{x}^*) d = \begin{pmatrix}
d_1 & d_2
\end{pmatrix}
\begin{pmatrix}
2 & 0 \\
0 & 2
\end{pmatrix}
\begin{pmatrix}
d_1 \\
d_2
\end{pmatrix} = 2d_1^2 + 2d_2^2
$$

Since
$$
2d_1^2 + 2d_2^2 > 0
$$

for any $d \neq 0 $, hence, the critical point $(x_1,x_2)=(0,0)$ is a **local minimum**.

### 4. Checking critical points with the eingenvalues of the $\nabla ^2 f(\mathbf{x}^*) $

If we check the eingevalues of $\nabla ^2 f(\mathbf{x}^*)$ we can also determine if the critical point is a local minimum or maximum. By solving the characteristic equation:

$$
\det(\nabla ^2 f(\mathbf{x}^*) - \lambda I) = 0.
$$

Now, we compute the determinant:

$$
\det(\nabla ^2 f(\mathbf{x}^*) - \lambda I) = (2 - \lambda)(2 - \lambda) = (2 - \lambda)^2
$$

Setting the determinant equal to zero:

$$
(2 - \lambda)^2 = 0
$$

This leads to:

$$
2 - \lambda = 0 \implies \lambda = 2
$$

Thus, the eigenvalues of $\nabla ^2 f(\mathbf{x}^*)$ are:

- $ \lambda_1 = 2 $
- $ \lambda_2 = 2 $

Since both eigenvalues are strictly positive, we conclude that there is a **local minimum** at $ x^* = (0, 0) $.

```python
import numpy as np

# Define the functions
def f1(x1, x2):
    return -x1**2 - x2**2

def f2(x1, x2):
    return x1**2 - x2**2

def f3(x1, x2):
    return x1**2

# Calculate the Hessian matrix
def hessian_f1(x1, x2):
    return np.array([[ -2,  0],
                     [  0, -2]])

def hessian_f2(x1, x2):
    return np.array([[ 2,  0],
                     [ 0, -2]])

def hessian_f3(x1, x2):
    return np.array([[ 2, 0],
                     [ 0, 0]])

# Analyze critical points
def analyze_critical_points():
    critical_points_info = {}

    # Points for f1
    crit_point_f1 = (0, 0)
    H_f1 = hessian_f1(*crit_point_f1)
    eigvals_f1 = np.linalg.eigvals(H_f1)
    critical_points_info['f1'] = (crit_point_f1, eigvals_f1)

    # Points for f2
    crit_point_f2 = (0, 0)
    H_f2 = hessian_f2(*crit_point_f2)
    eigvals_f2 = np.linalg.eigvals(H_f2)
    critical_points_info['f2'] = (crit_point_f2, eigvals_f2)

    # Points for f3
    crit_point_f3 = (0, 0)
    H_f3 = hessian_f3(*crit_point_f3)
    eigvals_f3 = np.linalg.eigvals(H_f3)
    critical_points_info['f3'] = (crit_point_f3, eigvals_f3)

    return critical_points_info

# Analyze the critical points
critical_points_info = analyze_critical_points()

# Print the results
for func, (point, eigvals) in critical_points_info.items():
    print(f"Function: {func}, Critical Point: {point}, Eigenvalues: {eigvals}")

    # Determine the nature of the critical point
    if np.all(eigvals > 0):
        print("  Nature: Local Minimum")
    elif np.all(eigvals < 0):
        print("  Nature: Local Maximum")
    else:
        print("  Nature: Saddle Point")
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x, y)
def f(x, y):
    return x**2*(4-2.1*x**2+1/3*x**4)+x*y+y**2*(-4+4*y**2)

# Generate x1 and x2 values over a grid for the 3D plot
x_vals = np.linspace(-2, 2, 400)
y_vals = np.linspace(-1, 1, 400)
x, y = np.meshgrid(x1_vals, x2_vals)

# Compute the function values
f_vals = f(x, y)

# Step 1: Plot the function as a 3D surface plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, f_vals, cmap='viridis', edgecolor='none')

# Set labels and title
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$f(x, y)$')

# Show the plot
plt.show()
```

### 2. Analytical Computation of the Gradient

Consider the function:

$$f(x,y)=x^2(4-2.1x^2+\frac{1}{3}x^4)+xy+y^2(-4+4y^2)$$

The gradient of a function $f(x, y)$ is given by the vector of partial derivatives:

$$\nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)$$

Let's compute the partial derivatives of $f(x, y)$:

$$
\frac{\partial f}{\partial x} = 8x+y-8.4x^3+2x^5$$

$$\frac{\partial f}{\partial y} = x-8y+16y^3$$

Thus, the gradient is:

$$ \nabla f(x, y) = \left( 8x+y-8.4x^3+2x^5, x-8y+16y^3 \right)
$$



### 3. Numerically compute an approximation of the points $x^*$ at which $\nabla f(x^∗) = 0$.

a) We compute $||\nabla f(x,y)||^2$ for all values $(x,y)\in [-2,2]\times[-1,1]$ with a step of $0,005$

```python
import numpy as np

# Define the function f(x, y)
def mod_nabla_f(x, y):
    return (8*x+y-8.4*x**3+2*x**5)**2+(x-8*y+16*y**3)**2

# Define the range and step size
x_values = np.arange(-2, 2.005, 0.005)  # x values from -2 to 2
y_values = np.arange(-1, 1.005, 0.005)  # y values from -1 to 1

# Create a meshgrid of x and y values
x_grid, y_grid = np.meshgrid(x_values, y_values)

# Apply the function to each (x, y) pair using vectorization
res = mod_nabla_f(x_grid, y_grid)
```

b) We find the values that are smaller than their 8 neightbors

```python
min_points = []

for i in range(1, res.shape[0]):
  for j in range(1, res.shape[1]):
    if(res[i,j]<res[i-1,j-1] and res[i,j]<res[i-1,j] and res[i,j]<res[i-1,j+1] and res[i,j]<res[i,j-1] and res[i,j]<res[i,j+1] and res[i,j]<res[i+1,j-1] and res[i,j]<res[i+1,j] and res[i,j]<res[i+1,j+1]):
      min_points.append((-1+0.005*i,-2+0.005*j))
```

c) We find these points, and the values of $||\nabla f(x,y)||^2$

```python
print("Values: ", min_points)

min_vals = []
for i, x in enumerate(min_points):
  min_vals.append(mod_nabla_f(x[1],x[0]))

print("Values at these points: ", min_vals)
```

### 4. Compute the analytical expression of the Hessian



The Hessian matrix is:
$$H(f(x,y)) = \begin{pmatrix} 8-25.2x^3+10x^4 & 1 \\ 1 & 48x^2 - 8 \end{pmatrix} $$

```python
import numpy as np

def hessian_f(x,y):
  return [8-25.2*x**3+10*x**4,1],[1,-8+48*y**2]

hessians = []
for i in min_points:
  hessians.append(hessian_f(i[0], i[1]))

eigvals = []
for i in hessians:
  eigvals.append(np.linalg.eigvals(i))

i = 0
for e in eigvals:
  if(e[0]>0 and e[1]>0):
    print("Point ", min_points[i], "is a minimum")
  elif(e[0]<0 and e[1]<0):
    print("Point ", min_points[i], "is a maximum")
  else:
    print("Point ", min_points[i], "is a saddle point")
  i+=1
```
